```python
import re

def space_to_snake(column_name):
    """Replace space with underscore """
    tmp = re.sub(r"\s+",'_',column_name)
    return tmp.lower()
```

```python
data.columns
```

```python
tmp = []
for col in data.columns:
    tmp.append(space_to_snake(col))
data.columns=tmp
    
    
```
